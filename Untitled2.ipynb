```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-5, 5, 0.001)

unit_step = (t >= 0).astype(float)
unit_impulse = (t == 0).astype(float)
ramp = t * (t >= 0).astype(float)
sine_wave = np.sin(2 * np.pi * 1 * t)
exponential = np.exp(-0.5 * t) * (t >= 0).astype(float)

fig, axs = plt.subplots(5, 1, figsize=(10, 10))

axs[0].plot(t, unit_step, linewidth=2, color='b')
axs[0].set_title('Unit Step Signal')
axs[0].grid(True)

axs[1].stem(t, unit_impulse, linefmt='r-', markerfmt='ro', basefmt=' ')
axs[1].set_title('Unit Impulse Signal')
axs[1].grid(True)

axs[2].plot(t, ramp, linewidth=2, color='g')
axs[2].set_title('Ramp Signal')
axs[2].grid(True)

axs[3].plot(t, sine_wave, linewidth=2, color='m')
axs[3].set_title('Sinusoidal Signal (1 Hz)')
axs[3].grid(True)

axs[4].plot(t, exponential, linewidth=2, color='c')
axs[4].set_title('Exponential Decay Signal')
axs[4].grid(True)

plt.tight_layout()
plt.show()
```
